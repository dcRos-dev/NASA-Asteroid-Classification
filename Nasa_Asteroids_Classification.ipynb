## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataframe

```python
df = pd.read_csv('./nasa.csv')
df.head(10)
```

```python
df.shape
```

## Checking for Null Values

```python
df.info()
```

## Dropping Useless Columns

```python
df.drop(['Neo Reference ID','Name','Orbit ID',
         'Close Approach Date','Epoch Date Close Approach',
         'Orbit Determination Date'],
        axis=1, inplace=True)

df.head()
```

## Data Transformation

```python
hazardous_labels = pd.get_dummies(df['Hazardous'], dtype=int)
hazardous_labels
```

## Converting Hazardous boolean values to 0,1

```python
df = pd.concat([df,hazardous_labels],axis=1)
df.info()
```

## Dropping Column Hazardous

```python
df = df.drop(['Hazardous'], axis=1)
df.info()
```

## Dropping Orbiting Body & Equinox

```python
df = df.drop(['Orbiting Body','Equinox'], axis=1)
```

## Checking Data Correlationships with heatmap

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot= True)
```

## Cleaning redundant Datas

```python
df = df.drop(['Est Dia in KM(max)', 'Est Dia in M(min)', 'Est Dia in M(max)', 'Est Dia in Miles(min)'
             ,'Est Dia in Miles(max)', 'Est Dia in Feet(min)', 'Est Dia in Feet(max)',
             'Relative Velocity km per hr', 'Miles per hour', 'Miss Dist.(lunar)',
             'Miss Dist.(kilometers)', 'Miss Dist.(miles)'], axis = 1)
df.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot= True)
```

## Only True columns is needed

```python
df.drop([False],axis=1, inplace=True)
```

```python
df.head()
```

## Model Bulding

```python
x = df.drop([True], axis = 1)
y = df[True].astype(int)
```

```python
from sklearn.model_selection import train_test_split
#Splitting between Train and Test
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=0, test_size=0.3)
```

```python
from xgboost import XGBClassifier
from matplotlib import pyplot
from xgboost import plot_importance

#Using XGBClassifier foor Gradient Boosting
xgb_model=XGBClassifier()
xgb_model.fit(x_train,y_train)
plot_importance(xgb_model)
pyplot.show()
```

```python
from sklearn.metrics import accuracy_score ,classification_report, confusion_matrix

predictions = xgb_model.predict(x_test)
acc=accuracy_score(y_test, predictions)
print(str(np.round(acc*100,2))+'%')
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```
